```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('students.csv')
df
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
## Droping the column Unnamed from table

df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
### Changing weekly study hours column

df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('05-oct', '05-10')
df.head()
```

```python
### Gender Distribution

plt.figure(figsize=(5,5))
ax =sns.countplot(data=df, x='Gender')
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
gb
```

```python
sns.heatmap(gb, annot=True)
plt.show()
```

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
gb1
```

```python
sns.heatmap(gb1, annot=True)
plt.show()
```

```python
df['EthnicGroup'].unique()
```

```python
### Distribution of Ethnic Group
groupA = df.loc[df['EthnicGroup'] == 'group A'].count()
groupB = df.loc[df['EthnicGroup'] == 'group B'].count()
groupC = df.loc[df['EthnicGroup'] == 'group C'].count()
groupD = df.loc[df['EthnicGroup'] == 'group D'].count()
groupE = df.loc[df['EthnicGroup'] == 'group E'].count()

mylist = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
plt.pie(mylist, labels=['group A', 'group B', 'group C', 'group D', 'group E'], autopct='%1.1f%%')
plt.title('Distribution of Ethnic Group')
plt.show()
```

```python
ax = sns.countplot(data=df, x='EthnicGroup')
ax.bar_label(ax.containers[0])
plt.title('Ethnic Group Distribution')
plt.show()
```
